```python
import pandas as pd
```

```python
FPL_df = pd.read_csv('epl_final.csv')
FPL_df.columns
```

```python
Chelsea_df = FPL_df[FPL_df['HomeTeam'] == 'Chelsea']
Chelsea_df
```

```python
Chelsea_Home_df = FPL_df[FPL_df['HomeTeam'] == 'Chelsea']
len(Chelsea_Home_df)
```

```python
Chelsea_Away_df = FPL_df[FPL_df['AwayTeam'] == 'Chelsea']
Chelsea_Away_df
```

```python
Chelsea_Home_Win_df = Chelsea_Home_df[Chelsea_Home_df['FullTimeResult'] == 'H']
Chelsea_Home_Draw_df = Chelsea_Home_df[Chelsea_Home_df['FullTimeResult'] == 'D']
Chelsea_Home_Loss_df = Chelsea_Home_df[Chelsea_Home_df['FullTimeResult'] == 'A']

len(Chelsea_Home_Loss_df)
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
Chelsea_Home_df.plot(kind='line', x='Season', y='FullTimeHomeGoals')
```

```python
Chelsea_Away_df = FPL_df[FPL_df['AwayTeam'] == 'Chelsea']

Chelsea_Away_Win_df = Chelsea_Away_df[Chelsea_Away_df['FullTimeResult'] == 'A']
Chelsea_Away_Draw_df = Chelsea_Away_df[Chelsea_Away_df['FullTimeResult'] == 'D']
Chelsea_Away_Loss_df = Chelsea_Away_df[Chelsea_Away_df['FullTimeResult'] == 'H']

len(Chelsea_Away_Win_df)
```

```python
Chelsea_df = Chelsea_Home_df.merge(Chelsea_Away_df, how='outer')
Chelsea_df
```

```python
Chelsea_Away_df.plot(kind='line', x='Season', y='FullTimeAwayGoals')
```

```python
Chelsea_Home_df.plot(kind='line', x='Season', y='FullTimeHomeGoals')
Chelsea_Away_df.plot(kind='line', x='Season', y='FullTimeAwayGoals')
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.subplots()

Chelsea_Home_df.plot(kind='line', x='Season', y='FullTimeHomeGoals', ax=ax)
Chelsea_Away_df.plot(kind='line', x='Season', y='FullTimeAwayGoals', ax=ax)

fig.show()
```
